Lyrics Scraper to collect dataset by artists. 

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
artist_name = "Orishas"
```

```python
def get_lyrics_from_artist_df(artist, num_songs = None):
    url = 'https://www.lyrics.com/artist/' + artist.split(' ')[0].lower()
    for i in range(1,len(artist.split())):
        url += '%20' + artist.split(' ')[i].lower()
        
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    
    son = []

    songs = soup.find_all('td', {'class': 'tal qx'})

    base_url = 'https://www.lyrics.com/'
    print("Number of Songs found for "+str(artist)+": "+str(len(songs)))
    lyr = []
    iterspace = len(songs)
    
    if num_songs:
        iterspace = min(num_songs, iterspace)
        
    for i in range(iterspace):
        print("\nProcessing: "+str(songs[i].text))
        if songs[i].a is None:
            pass
        else:
            lyr_url = base_url + songs[i].a.attrs['href']
            r = requests.get(lyr_url)
            son.append(songs[i].text)
            soup = BeautifulSoup(r.text, 'html.parser')
            lyr.append(soup.find('pre', {'id': 'lyric-body-text'}).text)
          

    df = pd.DataFrame({'song': son,  'lyrics': lyr})
    return df
```

```python
def lyrics_cleaning(x):
    fixed = ''
    lyr_lst = x.split('\r\n')
    
    for item in lyr_lst[:-1]:
        fixed += item + ' '
        
    fixed += lyr_lst[-1]
    return fixed
```

```python
df = get_lyrics_from_artist_df(artist_name)
```

```python
df = df.drop_duplicates()
```

```python
df['lyrics'] = df['lyrics'].apply(lyrics_cleaning)
lyrics = list(df['lyrics'])
```

```python
with open("data/orishas-lyrics.txt","w",encoding="utf-8") as file:
    for item in lyrics:
        file.write('%s\n' % item)
```
